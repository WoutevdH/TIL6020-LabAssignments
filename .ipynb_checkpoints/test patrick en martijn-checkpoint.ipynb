```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
```

```python
path = r"CBS.csv"

data = pd.read_csv(path,sep=';',skiprows=[0, 1, 2, 3 ],header=[0, 1, 2],index_col=[0])
data = data.replace(',', '.', regex=True)

data = data.dropna(thresh=data.shape[1] - 1)
data = data.drop('Unnamed: 1_level_0',axis=1,level=0)
data.index = data.index.str.replace('*', '', regex=False)
display(data)

```

```python

totaal = data.iloc[:,[0, 1, 2, 3]]
Bestuurder = data.iloc[:,[0, 4, 5, 6]]
Passagier = data.iloc[:,[0, 7, 8, 9]]
Trein = data.iloc[:,[0, 10, 11, 12]]
BTM = data.iloc[:,[0, 13, 14, 15]]
Fiets = data.iloc[:,[0, 16, 17, 18]]
Lopen = data.iloc[:,[0, 19, 20, 21]]
Overige = data.iloc[:,[0, 22, 23, 24]]


# List of the datasets
datasets = [totaal, Bestuurder, Passagier, Trein, BTM, Fiets, Lopen, Overige]

# Iterate over each dataset
for df in datasets:
    # Drop the first two levels of the column index (if applicable)
    df.columns = df.columns.droplevel([0, 1])
    
    # Use .loc to modify the DataFrame in place and avoid SettingWithCopyWarning
    df.loc[:, 'aantal'] = pd.to_numeric(df['aantal'], errors='coerce')
    df.loc[:, 'km'] = pd.to_numeric(df['km'], errors='coerce')
    df.loc[:, 'Minuten'] = pd.to_numeric(df['Minuten'], errors='coerce')
    #calculate speed
    df.loc[:, 'snelheid'] = df['km'] / (df['Minuten'] / 60)
```

**Hypothesis** 

```python
display(totaal)

import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'totaal' is already defined as per your structure
# Calculate average speed for each 'Reismotieven' across all years
average_speed = totaal.groupby('Reismotieven')['snelheid'].mean().reset_index()

# Sort the values for better visualization (optional)
average_speed = average_speed.sort_values(by='snelheid', ascending=False)

# Plotting the average speed by Reismotieven
plt.figure(figsize=(12, 6))
plt.barh(average_speed['Reismotieven'], average_speed['snelheid'], color='skyblue')  # Horizontal bar plot
plt.title('Average Speed by Travel Motivation')
plt.xlabel('Average Speed (km/h)')
plt.ylabel('Travel Motivation')
plt.grid(axis='x')
plt.tight_layout()
plt.show()

```

In the graph above, it is clearly visible that there is a big difference in average speed between different travel motives. The average speed of all travel motives is about 27 km/h. Shopping/doing groceries has the lowest average speed of about 22 km/h. This is logical, because a lot of people in the Netherlands do their groceries close by and often take the bike. If people do their groceries by car, they mainly stay on urban roads where there is a speed limit of 50km/h. The highest average speed is achieved when going to and from work. This speed is around 37 km/h. This is probably caused by the fact that a lot of people do not work in the same city as where they live. This results in a higher travel distance, which means people often take the car (and travel via highways) or the train. Furthermore, it can also be concluded that following education, courses and bringing your child to childcare (23 km/h) and travelling during free time (24 km/h) is lower than the total average, while the category 'others' is a little bit above average with 37 km/h.  

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'totaal' DataFrame is already defined and looks similar to the one described earlier

# # Display the DataFrame
# display(totaal)

# # Resetting the index to treat 'Perioden' as a regular column
totaal.reset_index(inplace=True)

# # Display the DataFrame's columns and the first few rows
# print(totaal.columns)
# display(totaal.head())

# Grouping by 'index' (which is now 'Perioden') and 'Reismotieven', then calculating the mean speed
grouped_speed = totaal.groupby(['index', 'Reismotieven'])['snelheid'].mean().reset_index()

# Print the grouped data to verify
print(grouped_speed)

# Create the line plot for speed changes over time for each motive
plt.figure(figsize=(14, 7))

# Loop through each unique travel motive to plot
for motive in grouped_speed['Reismotieven'].unique():
    motive_data = grouped_speed[grouped_speed['Reismotieven'] == motive]
    plt.plot(motive_data['index'], motive_data['snelheid'], marker='o', label=motive)

# Add titles and labels
plt.title('Change of Speed Over Time by Travel Motivation')
plt.xlabel('Year')
plt.ylabel('Average Speed (km/h)')
plt.xticks(rotation=45)
# plt.xticks(range(min(grouped_speed['index']), max(grouped_speed['index']) + 1))  # Ensuring all years are shown

# Add a legend
plt.legend(title='Travel Motivation', loc='upper left')  # You can adjust the location as needed

# Add a grid for better readability
plt.grid()

# Adjust layout to fit everything nicely
plt.tight_layout()

# Show the plot
plt.show()
```

In the graph above the change of travel speed per motive over the last 20 years is made visible. It is clear that for most of the motives the average speed has increased, however the category 'other' motives has seen a decrease in the past 20 years. It is not completely clear why this decrease in average speed has occured and it is hard to see why this has occuredjcause it is not clear what the definition others does exactly include. It at least contains movemvents for businnes meetings, but for the rest there is no clear answer.
All other travel motives have seen an increase in speeed over the last 20 years. Especially travel speed from and to work and the travel speed for shopping / doing groceries has significanlty increased. This can have multiple reasons. One of the reasons is probably that the general wealth has increased, which means more people do have access to cars. Besides this, also technological improvements have increased the travel speed. However, due to maximum speed this can not have a big influence on the average speed. Finally, also improvement in lowering traffic jams can have an influence on the overall travel speed.
